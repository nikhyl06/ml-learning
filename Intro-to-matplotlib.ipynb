## Matplotlib exaple workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = np.arange(-100,101, dtype=int)
y = np.power(np.arange(-100,101, dtype=int),3)

fig, ax = plt.subplots(figsize = (10,10))

ax.plot(x,y);

ax.set(title = "The graph of a cubic function",
      xlabel = "x-axis",
      ylabel = "y-axis");
```

### let's try Heart Disease

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
heart_disease['age'].plot.hist(bins=45);
```

```python
heart_disease.plot.hist(subplots=True, figsize=(10,30));
```

```python
 over_50 = heart_disease[heart_disease['age']>50]
```

```python
over_50
```

```python
## pyplot
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

## pyplot method

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
```

## OO method with pyplot

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);

ax.set_xlim([45,100]);
```

## OO method

```python
fig, (ax1,ax2) = plt.subplots(figsize=(10,12),
                             nrows=2,
                             ncols=1,
                             sharex=True)
scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['chol'],
                    c=over_50['target'])

ax1.set(title="Heart Disease with cholestrol",
      ylabel='cholestrol level')

ax1.legend(*scatter.legend_elements(), title='Target')
ax1.axhline(y=over_50['chol'].mean(), linestyle = '--', color='b');




scatter = ax2.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                    c=over_50['target'])

ax2.set(title="Heart Disease with Max Heart Rate",
      xlabel = 'age',
      ylabel='Max Heart Rate')

ax2.legend(*scatter.legend_elements(), title='Target')
ax2.axhline(y=over_50['thalach'].mean(), linestyle = '--', color='b');


fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold');
```
